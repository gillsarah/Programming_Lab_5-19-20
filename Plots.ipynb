# Plotting Example

```python
path = '/Users/Sarah/Documents/GitHub/Programming_Lab_5-19-20/data'
#for the HW, hint: os.getcwd() and os.path.join() 
#try these in more than one coding environment 
filenames = ['Affordable_Rental_Housing_Developments.csv', 'Chicago_SES.csv']
```

```python
dfs = []
for f in filenames:
    df = pd.read_csv(os.path.join(path, f))
    dfs.append(df) 

dfs[0].head()
```

```python
dfs[1].head()
```

```python
dfs[0].drop(columns = ['Community Area Name', 'Property Type',
       'Property Name', 'Address', 'Zip Code', 'Phone Number',
       'Management Company', 'X Coordinate', 'Y Coordinate',
       'Latitude', 'Longitude', 'Location'], inplace = True)
```

```python
units = dfs[0].groupby('Community Area Number').sum()
units.head()
```

```python
chi_housing = units.merge(dfs[1], on ='Community Area Number', how = 'outer')
```

```python
chi_housing.head()
chi_housing.columns
```

```python
x = chi_housing['PERCENT OF HOUSING CROWDED']
y = chi_housing['Units']

plt.plot(x, y, 'o')
plt.show() #don't need this for Jupyter, but do for other coding environments
```

```python
plt.scatter(x, y)
plt.show() #don't need this for Jupyter, but do for other coding environments
```

```python
size = chi_housing['PER CAPITA INCOME ']/1000
plt.scatter(x, y, size)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, y, 'o')
ax.set_ylabel('Affordable Housing Units Per Community Area')
ax.set_xlabel('Percent of Communty Area Housing that is Crowded')
plt.show()
```

```python
def housing_plot(x, xlable):
    fig, ax = plt.subplots(figsize=(6,6))
    ax.plot(x, chi_housing['Units'], 'o')
    ax.set_xlabel(xlable)
    ax.set_ylabel('Affordable Housing Units Per Community Area')
    ax.spines['right'].set_visible(False) 
    ax.spines['top'].set_visible(False)
    plt.show()
 
housing_plot(chi_housing['PERCENT OF HOUSING CROWDED'], 'Percent of Communty Area Housing that is Crowded')
```

```python
housing_plot(chi_housing['HARDSHIP INDEX'], 'Hardship Index')
```

```python
def housing_and_inc_plot(x, xlable, size):
    fig, ax = plt.subplots(figsize=(6,6))
    ax.scatter(x = x, y = chi_housing['Units'], s = size)
    ax.set_xlabel(xlable)
    ax.set_ylabel('Affordable Housing Units Per Community Area')
    ax.spines['right'].set_visible(False) 
    ax.spines['top'].set_visible(False)
    plt.show()
 
housing_and_inc_plot(chi_housing['PERCENT OF HOUSING CROWDED'], 
             'Percent of Communty Area Housing that is Crowded',
             chi_housing['PER CAPITA INCOME ']/1000)
```

```python
#revist to have a ledgand (this ledgand is more complicated than you'll likely need)
def housing_and_inc_plot(x, xlable, size, sizelable):
    fig, ax = plt.subplots(figsize=(6,6))
    ax.scatter(x = x, y = chi_housing['Units'], s = size)
    ax.set_xlabel(xlable)
    ax.set_ylabel('Affordable Housing Units Per Community Area')
    ax.spines['right'].set_visible(False) 
    ax.spines['top'].set_visible(False)
    for a_list in [10, 20, 30]:
        plt.scatter([], [], c='k', alpha=0.3, s=a_list,
                    label=str(a_list) + 'k')
    plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title=sizelable)
    plt.show()
 
housing_and_inc_plot(chi_housing['PERCENT OF HOUSING CROWDED'], 
                    'Percent of Communty Area Housing that is Crowded',
                    chi_housing['PER CAPITA INCOME ']/1000,
                    'Per-Capita Income')
```

```python
housing_and_inc_plot(chi_housing['PERCENT OF HOUSING CROWDED'], 
                    'Percent of Communty Area Housing that is Crowded',
                    chi_housing['HARDSHIP INDEX'],
                    'Hardship Index')
```

```python
def housing_and_inc_plot(x, xlable, size, sizelable, thousands = ''):
    fig, ax = plt.subplots(figsize=(6,6))
    ax.scatter(x = x, y = chi_housing['Units'], s = size)
    ax.set_xlabel(xlable)
    ax.set_ylabel('Affordable Housing Units Per Community Area')
    ax.spines['right'].set_visible(False) 
    ax.spines['top'].set_visible(False)
    for a_list in [10, 20, 30]:
        plt.scatter([], [], c='k', alpha=0.3, s=a_list,
                    label=str(a_list) + thousands)
    plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title=sizelable)
    plt.show()

housing_and_inc_plot(chi_housing['PERCENT OF HOUSING CROWDED'], 
                    'Percent of Communty Area Housing that is Crowded',
                    chi_housing['HARDSHIP INDEX'],
                    'Hardship Index')
```

```python
housing_and_inc_plot(chi_housing['PERCENT OF HOUSING CROWDED'], 
                    'Percent of Communty Area Housing that is Crowded',
                    chi_housing['PER CAPITA INCOME ']/1000,
                    'Per-Capita Income',
                    'k')
```
